```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
names = ["Judith", "Abel", "Tyson", "Martha", "Beverley", "David", "Anabel"]
estimated_insurance_costs = [1000.0, 2000.0, 3000.0, 4000.0, 5000.0, 6000.0, 7000.0]
actual_insurance_costs = [1100.0, 2200.0, 3300.0, 4400.0, 5500.0, 6600.0, 7700.0]

# Calculate average actual insurance cost
average_cost = sum(actual_insurance_costs) / len(actual_insurance_costs)

# === Visualization 1: Estimated vs. Actual Insurance Costs ===
x = np.arange(len(names))  # Label locations
width = 0.35  # Width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, estimated_insurance_costs, width, label='Estimated Costs', color='blue', alpha=0.7)
bars2 = ax.bar(x + width/2, actual_insurance_costs, width, label='Actual Costs', color='red', alpha=0.7)

# Add a horizontal line for average actual cost
ax.axhline(average_cost, color='green', linestyle='--', linewidth=2, label=f'Avg Cost: {average_cost:.2f}')

# Labels and title
ax.set_xlabel('Names')
ax.set_ylabel('Insurance Costs ($)')
ax.set_title('Estimated vs. Actual Insurance Costs')
ax.set_xticks(x)
ax.set_xticklabels(names)
ax.legend()

# Show the chart
plt.show()


# === Visualization 2: Actual Insurance Costs with Average Line ===
plt.figure(figsize=(10, 6))
bars = plt.bar(names, actual_insurance_costs, color='skyblue', alpha=0.7, label='Actual Insurance Costs')

# Add a horizontal line for the average cost
plt.axhline(average_cost, color='red', linestyle='--', linewidth=2, label=f'Avg Cost: {average_cost:.2f}')

# Labels and title
plt.xlabel('Names')
plt.ylabel('Actual Insurance Costs ($)')
plt.title('Actual Insurance Costs Per Person')
plt.legend()

# Display the values on top of the bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():.0f}',
             ha='center', va='bottom', fontsize=10, color='black')

# Show the visualization
plt.show()
```
